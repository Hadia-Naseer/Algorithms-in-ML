# 4.0 S-ALGORITHM FUNCTION

```python
def find_s_algorithm(dataset):
    hypothesis = []
    for i, example in enumerate(dataset):
        if example[-1] == "Yes":
            if not hypothesis:
                hypothesis = list(example[:-1])
                print(f"Initial hypothesis after processing positive example {i+1}: {hypothesis}")
            else:
                for j in range(len(hypothesis)):
                    if hypothesis[j] != example[j]:
                        hypothesis[j] = "?"
                print(f"Updated hypothesis after processing positive example {i+1}: {hypothesis}")
    return hypothesis
```

```python
final_hypothesis = find_s_algorithm(data.values)
print("Final Hypothesis:", final_hypothesis)
```

# 5.0 QUESTION AND ANSWERS



 1.  How does find S- algorithm update hypothesis ?
*   The Find-S algorithm updates its hypothesis by starting with the most specific hypothesis and generalizing it only when a positive example has an attribute value different from the current hypothesis, replacing that attribute with a wildcard "?". It ignores negative examples and only modifies the hypothesis to remain consistent with all positive training examples.

2. Differentiate between find S- Algorithm and candidate elimination?
*  Find-S finds only the most specific hypothesis using positive examples, ignoring negatives.
Candidate Elimination finds the entire version space, considering both positive and negative examples.
