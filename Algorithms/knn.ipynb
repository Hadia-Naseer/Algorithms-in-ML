```python
def knn_classify(data, new_point, k):

    distances = []
    for index, row in data.iterrows():
        point = [row['SWEETNESS '], row['CRUNCHINESS']] # Note the space after SWEETNESS
        dist = euclidean_distance(new_point, point)
        distances.append({'distance': dist, 'category': row['CATEGORY']})
    distances.sort(key=lambda x: x['distance'])
    k_nearest = distances[:k]
    category_counts = {}
    for item in k_nearest:
        category = item['category']
        category_counts[category] = category_counts.get(category, 0) + 1
    predicted_category = max(category_counts, key=category_counts.get)
    return predicted_category
```

```python

Tomato = [4, 5]
predicted_category = knn_classify(data,Tomato, k=3)
print(f"The predicted category for the new thing is: {predicted_category}")
```
